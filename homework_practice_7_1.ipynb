### 2.1 Домашнее задание начального уровня

Вычислите евклидово расстояние между кластером *c1* и кластером *c3*. Это расстояние получилось меньше, или больше чем расстояние между кластером *c1* и кластером *c2*?

Координаты центра для трёх кластеров:

* "красный" кластер $c_1 = [1.0, 1.5]$
* "оранжевый" кластер $c_2 = [-1.0, -0.5]$
* "зеленый" кластер $c_3 = [1.0, -0.5]$

![clustering_example](https://248006.selcdn.ru/public/DS.%20Block%202.%20M7/clustering_example.png)

```python
from scipy.spatial.distance import euclidean

# Задаем координаты центров кластеров
c1 = [1.0, 1.5]
c2 = [-1.0, -0.5]
c3 = [1.0, -0.5]

# Вычисляем расстояние между c1 и c3
dist_c1_c3 = euclidean(c1, c3)
print(f"Расстояние между кластерами c1 и c3: {dist_c1_c3:.4f}")

# Для сравнения вычисляем расстояние между c1 и c2 (уже известно из урока)
dist_c1_c2 = euclidean(c1, c2)
print(f"Расстояние между кластерами c1 и c2: {dist_c1_c2:.4f}")

# Сравниваем расстояния
if dist_c1_c3 < dist_c1_c2:
    print("Расстояние между c1 и c3 МЕНЬШЕ, чем между c1 и c2")
elif dist_c1_c3 > dist_c1_c2:
    print("Расстояние между c1 и c3 БОЛЬШЕ, чем между c1 и c2")
else:
    print("Расстояния равны")
```

Расстояние между кластерами c1 и c3 получилось меньше (2.0), чем расстояние между кластерами c1 и c2 (≈3.606).

Геометрическое объяснение:
Если посмотреть на картинку с кластерами:

Красный кластер (c1) находится в координатах (1.0, 1.5)

Зеленый кластер (c3) находится в координатах (1.0, -0.5)

Оранжевый кластер (c2) находится в координатах (-1.0, -0.5)

Расстояние между c1 и c3 составляет ровно 2 единицы (по вертикали, так как координата x у них одинаковая: 1.0).
Расстояние между c1 и c2 больше, так как они находятся дальше друг от друга как по горизонтали, так и по вертикали.
