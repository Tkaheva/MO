### Домашняя работа

Даны точки $x$ и значения в этих точках $y$. Нормализуйте $y$ с помощью *z-score* и постройте график зависимости нормализованных значений от $x$. Для графика используйте `.scatter(x,y_tansformed)`

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt

x = np.linspace(1,10,num=10)
y = np.array(
    [1.,  3.,  4.,  2., 10.,  5.,  5.,  2.,  5., 10.],
    dtype=np.float32
)

# Способ 1: Использование StandardScaler из sklearn
scaler = StandardScaler()
y_normalized = scaler.fit_transform(y.reshape(-1, 1)).flatten()

# Способ 2: Ручной расчет z-score (альтернативный вариант)
# y_mean = np.mean(y)
# y_std = np.std(y)
# y_normalized = (y - y_mean) / y_std

# Построение графика
plt.figure(figsize=(10, 6))
plt.scatter(x, y_normalized, color='blue', s=80, alpha=0.7, edgecolors='black', linewidth=1)

plt.xlabel('x', fontsize=12)
plt.ylabel('Нормализованные значения y', fontsize=12)
plt.title('Зависимость нормализованных значений y от x', fontsize=14)
plt.grid(True, alpha=0.3)

# Добавляем горизонтальные линии для среднего и стандартных отклонений
plt.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='Среднее (μ=0)')
plt.axhline(y=1, color='orange', linestyle='--', alpha=0.5, label='+1σ')
plt.axhline(y=-1, color='orange', linestyle='--', alpha=0.5, label='-1σ')

plt.legend()
plt.tight_layout()
plt.show()

# Вывод статистики
print("Исходные значения y:")
print(y)
print(f"\nСтатистика исходных данных:")
print(f"Среднее: {np.mean(y):.4f}")
print(f"Стандартное отклонение: {np.std(y):.4f}")

print(f"\nНормализованные значения y (z-score):")
print(y_normalized)
print(f"\nСтатистика нормализованных данных:")
print(f"Среднее: {np.mean(y_normalized):.6f}")
print(f"Стандартное отклонение: {np.std(y_normalized):.6f}")
```
