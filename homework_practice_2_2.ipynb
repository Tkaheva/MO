У вас есть два набора точек – истинные значения *y_true* и предсказанные значения *y_pred*

* для каждой точки из *y_true* постройте величину ошибки $e = y-\hat{y}$ – это называется **остатки регрессии**
* возведите ошибки в квадрат $e^2$
* постройте график ошибок – зависимость $e^2$ от $e$

```python
import numpy as np

y_pred_ = np.array([30.0, 25.03, 30.57, 28.61, 27.94, 25.26, 23.0, 19.54, 11.52, 18.92, 19.0, 21.59, 20.91, 19.55, 19.28, 19.3, 20.53, 16.91, 16.18, 18.41, 12.52, 17.67, 15.83, 13.81, 15.68, 13.39, 15.46, 14.71, 19.55, 20.88, 11.46, 18.06, 8.81, 14.28, 13.71, 23.81, 22.34, 23.11, 22.92, 31.36])
y_true_ = np.array([24.0, 21.6, 34.7, 33.4, 36.2, 28.7, 22.9, 27.1, 16.5, 18.9, 15.0, 18.9, 21.7, 20.4, 18.2, 19.9, 23.1, 17.5, 20.2, 18.2, 13.6, 19.6, 15.2, 14.5, 15.6, 13.9, 16.6, 14.8, 18.4, 21.0, 12.7, 14.5, 13.2, 13.1, 13.5, 18.9, 20.0, 21.0, 24.7, 30.8])
```

График должен получится вот такой:

```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline

# Данные
y_pred = np.array([30.0, 25.03, 30.57, 28.61, 27.94, 25.26, 23.0, 19.54, 11.52, 18.92, 19.0, 21.59, 20.91, 19.55, 19.28, 19.3, 20.53, 16.91, 16.18, 18.41, 12.52, 17.67, 15.83, 13.81, 15.68, 13.39, 15.46, 14.71, 19.55, 20.88, 11.46, 18.06, 8.81, 14.28, 13.71, 23.81, 22.34, 23.11, 22.92, 31.36])
y_true = np.array([24.0, 21.6, 34.7, 33.4, 36.2, 28.7, 22.9, 27.1, 16.5, 18.9, 15.0, 18.9, 21.7, 20.4, 18.2, 19.9, 23.1, 17.5, 20.2, 18.2, 13.6, 19.6, 15.2, 14.5, 15.6, 13.9, 16.6, 14.8, 18.4, 21.0, 12.7, 14.5, 13.2, 13.1, 13.5, 18.9, 20.0, 21.0, 24.7, 30.8])

# 1. Вычисляем остатки регрессии (ошибки)
errors = y_true - y_pred

# 2. Возводим ошибки в квадрат
squared_errors = errors ** 2

# 3. Строим график зависимости квадратов ошибок от ошибок
plt.figure(figsize=(10, 6))
plt.scatter(errors, squared_errors, alpha=0.7, s=60, color='blue', edgecolors='black', linewidth=0.5)

# Добавляем линию y = x^2 для сравнения
x_line = np.linspace(min(errors)-1, max(errors)+1, 100)
y_line = x_line ** 2
plt.plot(x_line, y_line, 'r--', alpha=0.7, label='y = x²')

plt.xlabel('Ошибка (e)', fontsize=12)
plt.ylabel('Квадрат ошибки (e²)', fontsize=12)
plt.title('Зависимость квадрата ошибки от ошибки', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Дополнительная информация
print("Статистика ошибок:")
print(f"Средняя ошибка: {np.mean(errors):.4f}")
print(f"Среднеквадратичная ошибка (MSE): {np.mean(squared_errors):.4f}")
print(f"Корень из MSE (RMSE): {np.sqrt(np.mean(squared_errors)):.4f}")
print(f"Минимальная ошибка: {np.min(errors):.4f}")
print(f"Максимальная ошибка: {np.max(errors):.4f}")
```
